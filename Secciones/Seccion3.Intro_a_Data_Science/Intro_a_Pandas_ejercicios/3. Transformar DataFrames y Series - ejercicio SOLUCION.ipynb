Consideremos estados demócratas aquellos donde los candidatos demócratas tuvieron más votos y estados republicanos aquellos donde fueron los republicanos los que obtuvieron más votos. ¿Cuales son demócratas y cuales republicanos?

*pista: una manera de averiguar los estados que son democratas y republicanos es con pivot_table, usando la funcion aggregadora `aggfunc="sum"`*

<hr>

```python
votos_por_partido = pd.pivot_table(votos, values=["votes"], index="state",
                                   columns="party", aggfunc="sum")["votes"].reset_index()
```

```python
estados_democratas = votos_por_partido[votos_por_partido.Democrat> votos_por_partido.Republican].state.unique()
estados_republicanos = votos_por_partido[votos_por_partido.Democrat < votos_por_partido.Republican].state.unique()
```

```python
votos_republicanos = votos[votos.party=="Republican"]
```

```python
ganador_por_estado_republicano = (votos_republicanos[votos_republicanos.state.isin(estados_republicanos)]
 .groupby(["state", "candidate"], as_index=False)["votes"]
 .sum()
.sort_values(by=["state", "votes"], ascending=[True, False])
.groupby("state", as_index=False).head(1)
)
ganador_por_estado_republicano
```
